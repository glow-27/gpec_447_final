# Median income preprocessing

```python
import numpy as np
import pandas as pd
import janitor
import geopandas as gpd
```

```python
# change this according to which year's data you want
census_year = 2020

if census_year == 2020:
    income = pd.read_csv("data/2020_med_income/ACSST5Y2020.S1901-Data.csv")
    inc_col = 'S1901_C01_012E'
elif census_year == 2010:
    income = pd.read_csv("data/2010_med_income/ACSST5Y2010.S1903-Data.csv")
    inc_col = 'S1903_C02_001E'

income.head()
```

```python
print(np.sum(income[inc_col] == "-"))
income[inc_col] = np.where(income[inc_col] == "-", None, income[inc_col])
income[inc_col]
```

```python
# select the overall median estimate S1903_C02_001E
med_income = income[['GEO_ID', 'NAME', inc_col]]
med_income = med_income.iloc[1:]

med_income['tract'] = med_income['NAME'].str.extract(r"Census Tract ([0-9.]+)")
med_income['tract'] = med_income['tract'].astype(float).astype(str)
med_income.drop(columns=['NAME'], inplace=True)
med_income.rename(columns={inc_col: "med_inc"}, inplace=True)
med_income.head()
```

```python
# check for census tracts do not have an estimated median income?
np.sum(med_income['med_inc'].isna())
```
